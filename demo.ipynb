```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Set the time axis
t = np.linspace(0, 100, 1000)

# Generate a square wave
pulse_train = signal.square(2 * np.pi * 0.1 * t)  # Adjust frequency by changing 0.1

# Plot the square wave
plt.plot(t, pulse_train)
plt.ylim([-1.5, 1.5])
plt.grid(True)
plt.show()
```
